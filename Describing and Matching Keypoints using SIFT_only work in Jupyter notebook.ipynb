```python
#%matplotlib notebook
#%matplotlib 
import cv2
# import matplotlib
# matplotlib.use('nbagg')
import matplotlib.pyplot as plt
```

# Generated Image

```python
# reading the image
img1 = cv2.imread('original.png')
# convert to greyscale
#image = img.astype('uint8')
gray1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
```

```python
# create SIFT feature extractor
sift1 = cv2.xfeatures2d.SIFT_create()
```

```python
# detect features from the image
keypoints1, descriptors1 = sift1.detectAndCompute(img1, None)
```

```python
# draw the detected key points
sift_image1 = cv2.drawKeypoints(gray1, keypoints1, img1)
# # show the image
# cv2.imshow('image', sift_image)
# # save the image
cv2.imwrite("original_keypoints.png", sift_image1)
# cv2.waitKey(0)
# cv2.destroyAllWindows()

plt.imshow(sift_image1)
```

```python
descriptors1
```

```python
cv2.imwrite("original_descriptors.png", descriptors1)
plt.imshow(descriptors1)
```

```python
import cv2

# read the images
img1 = cv2.imread('well1_pic_line.png')  
img2 = cv2.imread('well2_pic_line (1).png')
# convert images to grayscale
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
# create SIFT object
sift = cv2.xfeatures2d.SIFT_create()
# detect SIFT features in both images
keypoints_1, descriptors_1 = sift.detectAndCompute(img1,None)
keypoints_2, descriptors_2 = sift.detectAndCompute(img2,None)
```

```python
# create feature matcher
bf = cv2.BFMatcher(cv2.NORM_L1, crossCheck=True)
# match descriptors of both images
matches = bf.match(descriptors_1,descriptors_2)
```

```python
# sort matches by distance
matches = sorted(matches, key = lambda x:x.distance)
# draw first 50 matches
matched_img = cv2.drawMatches(img1, keypoints_1, img2, keypoints_2, matches[:50], img2, flags=2)
```

```python
# save the image
cv2.imwrite("matched_images.jpg", matched_img)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
# show the image
%matplotlib notebook
plt.imshow(matched_img)
```
