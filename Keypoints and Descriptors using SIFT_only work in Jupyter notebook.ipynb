```python
import cv2
import matplotlib.pyplot as plt
```

# Generated Image

```python
# reading the image
img1 = cv2.imread('original.png')
# convert to greyscale
#image = img.astype('uint8')
gray1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
```

```python
# create SIFT feature extractor
sift1 = cv2.xfeatures2d.SIFT_create()
```

```python
# detect features from the image
keypoints1, descriptors1 = sift1.detectAndCompute(img1, None)
```

```python
# draw the detected key points
sift_image1 = cv2.drawKeypoints(gray1, keypoints1, img1)
# # show the image
# cv2.imshow('image', sift_image)
# # save the image
cv2.imwrite("original_keypoints.png", sift_image1)
# cv2.waitKey(0)
# cv2.destroyAllWindows()

plt.imshow(sift_image1)
```

```python
descriptors1
```

```python
cv2.imwrite("original_descriptors.png", descriptors1)
plt.imshow(descriptors1)
```
